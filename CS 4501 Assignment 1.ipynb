```python
whole_sp = jst_sp.drop([i for i in mis_cols], axis = 1)
#2375 columns total of which 176 missing values
#calculate PCA for most important terms

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
pca = PCA(n_components = 3)#tested and found this is where drop off is

locs = {'open': 0, 'high': 1, 'low': 2, 'close': 3, 'volume': 4}
for i in mis_cols:
    cl = re.findall('[a-z]+',i)#regex to only get high/low... as predictors
    cl = locs[cl[0]]
    sep = jst_sp.iloc[:, jst_sp.columns == i].isna()
    #y_fill = jst_sp.iloc[sep.iloc[:,0].values, jst_sp.columns == i]
    y_train = jst_sp.iloc[[not i for i in sep.values], jst_sp.columns == i]
    x_train = whole_sp.iloc[[not i for i in sep.values], cl::5]
    #x_fill = whole_sp.iloc[sep.iloc[:,0].values, cl::5]
    standardized = StandardScaler().fit_transform(whole_sp.iloc[:, cl::5])
    components = pca.fit_transform(standardized)#trying to find pca for df
    x = np.ones((len(sep)-sum(sep.values)[0], 4))
    x[:,:-1] = components[[not i for i in sep.values],:]#only calculate x for training set
    theta = np.linalg.inv(x.T.dot(x)).dot(x.T).dot(y_train)
    x_fill = np.ones((sum(sep.values)[0],4))
    x_fill[:,:-1] = components[sep.iloc[:,0].values]
    jst_sp.iloc[sep.iloc[:,0].values, jst_sp.columns == i] = x_fill.dot(theta)
```
